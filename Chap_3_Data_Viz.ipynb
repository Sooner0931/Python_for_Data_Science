# 2 seaborn

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
tips_df.tail()
```

```python
tips_df.describe()
```

```python
type(tips_df)
```

### 1: distplot (distribution plot)

```python
sns.distplot(tips_df['total_bill'])
```

```python
sns.distplot(tips_df['total_bill'], bins = 30)
```

```python
sns.distplot(tips_df['total_bill'], bins = 30, kde = False)
```

```python
sns.distplot(tips_df['total_bill'], bins = 100, kde = False)
```

### 2: relplot (relation plot)

```python
sns.relplot(x = 'tip', y = 'total_bill', data = tips_df)
```

```python
tips_df.head()
```

```python
sns.relplot(x = 'tip', y = 'total_bill', data = tips_df, hue = 'sex')
```

```python
sns.relplot(x = 'tip', y = 'total_bill', data = tips_df, hue = 'smoker')
```

```python
sns.relplot(x = 'tip', y = 'total_bill', data = tips_df, hue = 'size')
```

```python
sns.relplot(x = 'tip', y = 'total_bill', data = tips_df, hue = 'day')
```

```python
sns.relplot(x = 'tip', y = 'total_bill', data = tips_df, hue = 'time')
```

### 3: pairplot (relation plot extended)

```python
sns.pairplot(tips_df, hue = 'sex')
```

```python
sns.pairplot(tips_df, hue = 'sex', palette = 'coolwarm' )
```

### 4: countplot

```python
sns.countplot(x = 'sex', data = tips_df)
```
